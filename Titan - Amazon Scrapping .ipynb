```python
data_asin=[]
response=getAmazonSearch('titan+raga+watch+for+women')
#beautifulsoup is a package
soup=BeautifulSoup(response.content)
for i in soup.findAll("div",{'class':"sg-col-4-of-24 sg-col-4-of-12 sg-col-4-of-36 s-result-item sg-col-4-of-28 sg-col-4-of-16 sg-col sg-col-4-of-20 sg-col-4-of-32"}):
    data_asin.append(i['data-asin'])
```

```python
def footer(search_query):
    url="https://www.amazon.in/dp/"+search_query
    print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
    #print(page.content)
```

```python
link=[]
for i in range(len(data_asin)-1):
    response=footer(data_asin[i+1])
#beautifulsoup is a package
    soup=BeautifulSoup(response.content)
    for i in soup.findAll("a",{'data-hook':"see-all-reviews-link-foot"}):
        link.append(i['href'])
```

```python
def revi(search_query):
    url="https://www.amazon.in"+search_query
    #print(url)
    page=requests.get(url,cookies=cookie,headers=header)
    if page.status_code==200:
        return page
    else:
        return "Error"
```

```python
review=[]
for j in range(len(link)):
    for k in range(25):
        response=revi(link[j]+'&pageNumber='+str(k))
#beautifulsoup is a package
        soup=BeautifulSoup(response.content)
        for i in soup.findAll("span",{'data-hook':"review-body"}):
            review.append(i.text)
```
